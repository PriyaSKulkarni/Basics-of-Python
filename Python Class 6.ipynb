Write a function which gives the square of the number:

```python
def square(n):
  return (n**2)
```

```python
numbers = [10,20,30]
lst =[]
for i in numbers:
  lst.append(square(i))
print(lst)
```

Alternative solution:

```python
def square_n(lst):
  output =[]
  for i in lst:
    res = i ** 2
    output.append(res)
  return output
```

```python
square_n(numbers)
```

```python
def square_num(lst):
  return [i**2 for i in lst]
```

Positional and Keyword

```python
def average(num1=10,num2=20):
  avg = (num1 +num2)/2
  return avg

print(average())
print(average(40))
print(average(100,200))
print(average(num2 =70, num1=40))
print(average(num2 = 80))
```

```python
def average(num1,num2=20): # num1 is positional and num2 is keyword
  avg = (num1 +num2)/2
  return avg

#print(average()) # print error
print(average(40))
print(average(100,200))
print(average(num2 =70, num1=40))
print(average(num1 = 80))
```
