```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
path ="/content/drive/MyDrive/Pythonclass/Dataset/ga_edgelist.csv"
```

```python
df=pd.read_csv(path)
df.head()
```

```python
graph = nx.from_pandas_edgelist(df,source='from',target='to')
```

```python
pos = nx.circular_layout(graph)
pos = nx.random_layout(graph)
```

```python
nx.draw(graph,pos,node_color='red', with_labels=True, node_size = 1000)
plt.show()
```
